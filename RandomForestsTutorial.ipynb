###### First method that removes outliers: remove data that is further than m std from the mean

```python
def rmv_outliers(data, m=2):
    return data[abs(data - np.mean(data)) < m * np.std(data)]

clean_data = rmv_outliers(data, 1.5)
```

##### Second method of re

```python
min_p = 25
max_p = 75

alcohol = data.alcohol.values
print("Initial size", alcohol.size)

p = np.percentile(alcohol, [min_p, max_p])

diff = abs(p[1] - p[0])
cst = diff * 1.5
Q1 = p[0] - cst
Q3 = p[1] + cst
clean_data2 = alcohol[(Q1<alcohol) & (Q3>alcohol)]

print("clean data size: ", clean_data2.size)
clean_data2
```
